```python
'''
File rosalind_ini4.txt

Given: Two positive integers a and b (a<b<10000).

Return: The sum of all odd integers from a through b, inclusively.

You can use a % 2 == 1 to test if a is odd.

Sample Dataset: 
100 
200

Sample Output:
7500
______
4049 
8099
'''

a = int(input())
b = int(input())
sm = 0

for i in range(a, b+1):
    if not i % 2 == 0:
        sm += i
        
print(sm)
```

```python
'''
File rosalind_ini6.txt

Given: A string s of length at most 10000 letters.

Return: The number of occurrences of each word in s, where words are separated by spaces. Words are 
case-sensitive, and the lines in the output can be in any order.

Sample Dataset:
We tried list and we tried dicts also we tried Zen

Sample Output:
and 1
We 1
tried 3
dicts 1
list 1
we 2
also 1
Zen 1
'''

l = []
res = []

for i in input().split():
    l.append(i)

count = 0
    
for i in range(len(l)):
    for j in range(len(l)):
        if l[i] == l[j]:
            count += 1
    var = l[i] + ' ' + str(count)
    res.append(var)
    count = 0

print(*set(res), sep='\n')
```

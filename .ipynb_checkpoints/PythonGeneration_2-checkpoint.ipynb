```python
'''
На easy. На вход программе подаются два целых числа a и b. Напишите программу, которая выводит:

- сумму чисел a и b;
- разность чисел a и b;
- произведение чисел a и b;
- частное чисел a и b;
- целую часть от деления числа a на b;
- остаток от деления числа a на b;
- корень квадратный из суммы их 10-х степеней: sqrt(a**10+b**10)

Формат входных данных:
На вход программе подаются два целых числа a и b (b≠0), каждое на отдельной строке.

Формат выходных данных:
Программа должна вывести результаты математических операций в соответствии с условием задачи.
'''
import math

def easy(A, B):
    result = []
    if not B == 0:
        result.append(f'{A} + {B} = {A+B}')
        result.append(f'{A} - {B} = {A-B}')
        result.append(f'{A} * {B} = {A*B}')
        result.append(f'{A} / {B} = {A/B}')
        result.append(f'{A} // {B} = {A//B}')
        result.append(f'{A} % {B} = {A%B}')
        result.append(f'sqrt({A}**10+{B}**10) = {math.sqrt(A**10+B**10)}')
    return result

print(*easy(int(input('Введите первое число: ')), int(input('Введите второе число: '))), sep='\n')
```

```python
'''
Индекс массы тела. Напишите программу для вычисления и оценки индекса массы тела (ИМТ) человека. ИМТ 
показывает весит человек больше или меньше нормы для своего роста. ИМТ человека рассчитывают по формуле:

ИМТ = масса(кг) / (рост(м)*рост(м)), где масса измеряется в килограммах, а рост — в метрах.

- Масса человека считается оптимальной, если его ИМТ находится между 18.5 и 25. 
- Если ИМТ меньше 18.5, то считается, что человек весит ниже нормы. 
- Если значение ИМТ больше 25, то считается, что человек весит больше нормы.

Программа должна вывести:
- "Оптимальная масса", если ИМТ находится между 18.5 и 25 (включительно). 
- "Недостаточная масса", если ИМТ меньше 18.5
- "Избыточная масса", если значение ИМТ больше 25.

Формат входных данных: На вход программе подается два числа: масса и рост человека, каждое на отдельной 
строке. Все входные числа являются вещественными, используйте для них тип данных float.

Формат выходных данных: Программа должна вывести текст в соответствии с условием задачи.
'''

def BMI(weight, height):
    result = weight/height**2
    if result > 25:
        return 'Избыточная масса. ИМТ =', result, 'кг/м**2'
    elif result < 18.5:
        return 'Оптимальная масса. ИМТ =', result, 'кг/м**2'
    else:
        return 'Недостаточная масса. ИМТ =', result, 'кг/м**2'
    
print(*BMI(float(input('Введите свой вес (кг): ')), float(input('Введите свой рост (м): '))))
```

```python
'''
Стоимость строки. Дана строка текста. Напишите программу для подсчета стоимости строки, исходя из того, что
один любой символ (в том числе пробел) стоит 60 копеек. Ответ дайте в рублях и копейках в соответствии с 
примерами. В 1 рубле 100 копеек.

Формат входных данных: На вход программе подается строка текста.
Формат выходных данных: Программа должна вывести стоимость строки.

Sample Input 1:    Привет, как дела?!
Sample Output 1:   10 р. 80 коп.
'''

def text_cost(text):
    cost = 0
    text = [c.split() for c in text]
    for i in text:
        cost += 60
    return f'Стоимость текста: {cost//100} р. {cost%100} коп.'

print(text_cost(input('Введите текст: ')))
```

```python
'''
Переворот числа. Дано пятизначное или шестизначное натуральное число. Напишите программу, которая изменит 
порядок его последних пяти цифр на обратный.

Формат входных данных: На вход программе подается одно натуральное пятизначное или шестизначное число.
Формат выходных данных: Программа должна вывести число, которое получится в результате разворота, указанного
в условии задачи. Число нужно выводить без незначащих нулей.
'''

def number_transposition(number):
    return number[:-5] + number[-5:][::-1]
    
print(int(number_transposition(input())))
```

```python
'''
Standard American Convention. На вход программе подаётся натуральное число. Напишите программу, которая 
вставляет в заданное число запятые в соответствии со стандартным американским соглашением о запятых в больших
числах.

Формат входных данных: На вход программе подаётся натуральное число n (0 < n < 10**100).
Формат выходных данных: Программа должна вывести число с запятыми в соответствии с условием задачи.
'''

def SAC(number):
    if int(number) > 0 and int(number) < 10**100:   # проверяем условие задачи
        if len(number) <= 3:                        # если длина числа меньше или равна 3,то
            return number                           # сразу ее выводим без изменений
        else:                                       # если длина числа больше 3
            number = number[::-1]                   # переворачиваем число (1234 в 4321)
            count = 0                               # задаем счетчик на ноль    
            result = []                             # создаем пустой список в котором будет результат
            for i in range(len(number)):            # проходим циклом по индексам числа
                count += 1                          # считаем сколько цифр уже пройденно в цикле
                result.append(number[i])            # добавляем каждую цифру в результирующий список
                if count == 3:                      # проверяем, если счетчик насчитал уже 3 цифры числа
                    result.append(',')              # то добавляем в результирующий список запятую
                    count = 0                       # и обнуляем счетчик, потом цикл повторяется до последней ц
            if result[::1][-1] == ',':              # проверяем последний индекс результирующего списка, если
                result = result[::-1]               # он равен запятой, переворачиваем список
                result = result[1:]                 # делаем срез с первого индекса (удаляем нулевой с запятой)
                return result                       # возвращаем результирующий список
        return result[::-1]                         # если последний индекс не равен запятой, возвращаем рез с

print(*SAC(input()), sep='')
```

```python
# Standard American Convention. or

# list.pop([i]) - Возвращает элемент [на указанной позиции], удаляя его из списка.

def SAC(number):
    s = ''
    
    while len(number) >= 3:
        # с number.pop(-1) отрываем 3 последние цифры и ставим после них запятую:
        s += number.pop(-1) + number.pop(-1) + number.pop(-1) + ','
    
    s += ''.join(number[::-1])     # обрабатываем цифры, которые могли остаться (их 1 или 2)
    s = s[::-1]                    # переворачивааем результат в первоначальный вид
    s = s.lstrip(',')              # убираем лишнюю запятую (стоит в самом начале)
    
    return s                             

print(*SAC(list(input())), sep='')
```

```python
# Standard American Convention. or


def SAC(number):
    number = number[::-1]                                    # переворачиваем число
    count = 0                                                # задаем нулевой счетчик
    for i in range(3, len(number), 3):                       # проходим циклом от 3 до длины числа с шагом 3
        number = number[:i+count] + ',' + number[i+count:]   # расставляем запятые после каждых 3 цифр
        count += 1                                           # увеличиваем счетчик на каждой итерации
    return number[::-1]                                      # выводим результат, предварительно перевернув обр

print(*SAC(input()), sep='')
```

```python
'''
Задача Иосифа Флавия. n человек, пронумерованных числами от 1 до n, стоят в кругу. Они начинают считаться,
каждый k-й по счету человек выбывает из круга, после чего счет продолжается со следующего за ним человека. 
Напишите программу, определяющую номер человека, который останется в кругу последним.

Формат входных данных: На вход программе подаются два числа n и k, записанные на отдельных строках.
Формат выходных данных: Программа должна вывести одно число – номер человека, который останется в кругу 
последним.

Примечание 1. Задача Иосифа Флавия — задача, входящая в одну из ранних работ по занимательной математике 
(1612 года) Баше де Мезириака. Задача заключается в следующем: по кругу стоит 41 воин, начиная с первого 
воина они убивают каждого третьего. Спрашивается, в каком месте нужно встать, чтобы остаться последним 
выжившим. В более общей формулировке участвует n воинов, которые считаются по кругу, и убивают каждого k-го. 
Название задачи восходит к истории, случившейся с Иосифом Флавием во время Иудейской войны.

Примечание 2. Визуализацию работы алгоритма можно посмотреть тут: https://www.geogebra.org/m/ExvvrBbR

Примечание 3. Подробный разбор математического алгоритма нахождения ответа: 
https://drive.google.com/file/d/1XFywConyqrFRvcSo36frn2T34Mrxede8/view
'''

def josephus_problem(n, k):
    n = [i for i in range(1, n+1)] # создаем список от 1 до n+1
    
    while len(n) > 1:            # пока в списке не останется лишь одно число,
        for i in range(k-1):     # иди в цикл, где перебираются элементы от 0 до k-1
            n.append(n[0])       # добавь в конец списка самый первый элемент этого же списка
            del n[0]             # теперь удали этот первый элемент из списка
        del n[0]                 # удали первый эдемент списка, когда выйдешь из цикла (останется лишь 1 число)
    return n                     # выведи результат

print(*josephus_problem(int(input()), int(input())))
```

```python
# Задача Иосифа Флавия. or

def josephus_problem(n, k):
    result = 0
    
    for i in range(1, n+1):
        result = (result + k) % i
    return result + 1

print(josephus_problem(int(input()), int(input())))
```

```python
'''
Координатные четверти. Дан набор точек на координатной плоскости. Необходимо подсчитать и вывести количество
точек, лежащих в каждой координатной четверти.

Формат входных данных: В первой строке записано количество точек. Каждая следующая строка состоит из двух 
целых чисел — координат точки (сначала абсцисса – x, затем ордината – y), разделенных символом пробела.
Формат выходных данных: Программа должна вывести количество точек, лежащих в каждой координатной четверти, 
как в примерах.

Примечание. Учтите, что точки, лежащие на осях координат, не принято относить к какой-либо координатной 
четверти.
'''

def coordinate_system(number):
    
    l = []                                     # заводим пустой список в котором будут координаты x и y
    x = ''                                     # заводим пустую строку, в которой будут координату x
    y = ''                                     # заводим пустую строку, в которой будут координаты y
    counter = [0, 0, 0, 0]                     # заводим список-счетчик, в котором будет результаты из цикла 
    result = []                                # заводим пустой список в котором будет результат
    
    print('\n')
    
    for i in range(number):                    # цикл для ввода данных ровно number раз
        l += [int(n) for n in input(f'Введите координаты {i+1} точки через 1 пробел: ').split()]
            
    x = l[::2]                                 # из обшего списка берутся все координаты x
    y = l[1::2]                                # из общего списка берутся все координаты y
    
    print('\n')
    
    print(f'{l} - общий список координат введенных точек', end='\n')
    print(f'{x} - координаты x', end='\n')
    print(f'{y} - координаты y', end='\n') 
    
    print('\n')
    
    for i in range(number):                    # определяем четверть по координатам , кол-во итераций = number
        if x[i] > 0 and y[i] > 0:              # берем i-тый элемент из списка x и y, проверяем условие
            counter[0] += 1                    # счетчик для первой четверти, индекс 0
        elif x[i] < 0 and y[i] > 0:            # берем i-тый элемент из списка x и y, проверяем условие
            counter[1] += 1                    # счетчик для второй четверти, индекс 1
        elif x[i] < 0 and y[i] < 0:            # берем i-тый элемент из списка x и y, проверяем условие
            counter[2] += 1                    # счетчик для третьей четверти, индекс 2
        elif x[i] > 0 and y[i] < 0:            # берем i-тый элемент из списка x и y, проверяем условие
            counter[3] += 1                    # счетчик для четвертой четверти, индекс 3 

    result = [f'Количество точек в первой четверти: {counter[0]}', 
              f'Количество точек во второй четверти: {counter[1]}',
              f'Количество точек в третьей четверти: {counter[2]}', 
              f'Количество точек в четвертой четверти: {counter[3]}']
    
    return result

print(*coordinate_system(int(input('Введите количество точек: '))), sep='\n')
```

```python
# Координатные четверти. or Проверка условий сразу после ввода первых координат

def coordinate_system(number):
    
    counter = [0, 0, 0, 0]                     # заводим список-счетчик, в котором будет результаты из цикла 
    result = []                                # заводим пустой список в котором будет результат
    
    print('\n')
    
    for i in range(number):                    # цикл для ввода данных ровно number раз
        x, y = [int(n) for n in input(f'Введите координаты {i+1} точки через 1 пробел: ').split()]
        
        if x > 0 and y > 0:                    # проверяем условие
            counter[0] += 1                    # счетчик для первой четверти, индекс 0
        elif x < 0 and y > 0:                  # проверяем условие
            counter[1] += 1                    # счетчик для второй четверти, индекс 1
        elif x < 0 and y < 0:                  # проверяем условие
            counter[2] += 1                    # счетчик для третьей четверти, индекс 2
        elif x > 0 and y < 0:                  # проверяем условие
            counter[3] += 1                    # счетчик для четвертой четверти, индекс 3 

    result = [f'Количество точек в первой четверти: {counter[0]}', 
              f'Количество точек во второй четверти: {counter[1]}',
              f'Количество точек в третьей четверти: {counter[2]}', 
              f'Количество точек в четвертой четверти: {counter[3]}']
    
    print('\n')
    
    return result

print(*coordinate_system(int(input('Введите количество точек: '))), sep='\n')
```

```python
'''
Больше предыдущего. На вход программе подается строка текста из натуральных чисел. Из неё формируется список
чисел. Напишите программу подсчета количества чисел, которые больше предшествующего им в этом списке числа. 

Формат входных данных: На вход программе подается строка текста из разделенных пробелами натуральных чисел.
Формат выходных данных: Программа должна вывести одно число – количество элементов списка, больших предыдущего.

Примечание. Разберём первый тест. У нас формируется список [1, 2, 3, 4, 5], в нём 2>1, 3>2, 4>3, 5>4. 
То есть получается 4 таких случая, где текущее число больше предыдущего. Поэтому ответ 4.
'''

def higher_previous_number(number):
    number = [int(n) for n in number.split()]
    counter = 0
    
    for i in range(1, len(number)):
        if number[i-1] < number[i]:
            counter += 1
    
    return counter

print(higher_previous_number(input('Введите строку текста из натуральных чисел: ')))
```

```python
'''
Назад, вперёд и наоборот. На вход программе подается строка текста из натуральных чисел. Из элементов строки
формируется список чисел. Напишите программу, которая меняет местами соседние элементы списка 
(a[0] c a[1], a[2] c a[3] и т.д.). Если в списке нечетное количество элементов, то последний остается на 
своем месте.

Формат входных данных: На вход программе подается строка текста, содержащая натуральные числа, разделенные 
пробелами.
Формат выходных данных: Программа должна вывести измененный список, разделяя его элементы одним пробелом.
'''

def element_transfer(number):
    number = [int(n) for n in number.split()]
    
    for i in range(0, len(number)-1, 2):
        number[i], number[i+1] = number[i+1], number[i]
        
    return number

print(*element_transfer(input('Введите строку текста с натуральными числами, разделенными 1 пробелом: ')))
```

```python
'''
Сдвиг в развитии. На вход программе подается строка текста из натуральных чисел. Из элементов строки 
формируется список чисел. Напишите программу циклического сдвига элементов списка направо, когда последний 
элемент становится первым, а остальные сдвигаются на одну позицию вперед, в сторону увеличения индексов.

Формат входных данных: На вход программе подается строка текста из разделенных пробелами натуральных чисел.
Формат выходных данных: Программа должна вывести элементы измененного списка с циклическим сдвигом, разделяя 
его элементы одним пробелом.
'''

def displacement(number):
    number = [int(n) for n in number.split()]
    result = [number[-1]] + number[:-1]
    return result

print(*displacement(input('Введите строку текста с натуральными числами, разделенными 1 пробелом: ')))
```

```python
'''
Различные элементы. На вход программе подается строка текста, содержащая натуральные числа, расположенные по 
неубыванию. Из строки формируется список чисел. Напишите программу для подсчета количества разных элементов в
списке.

Формат входных данных: На вход программе подается строка текста, содержащая натуральные числа, разделенные 
символом пробела, расположенные по неубыванию.
Формат выходных данных: Программа должна вывести одно число – количество различных элементов списка.

Примечание. Задачу можно решить без множеств.
'''

def various_elements(number):
    number = [int(n) for n in number.split()]
    result = []
    
    for i in number:
        if not i in result:
            result.append(i)
    
    return len(result)

print(various_elements(input('Введите строку текста с натуральными числами, разделенными 1 пробелом: ')))
```

```python
'''
Произведение чисел. Напишите программу для определения, является ли число произведением двух чисел из данного
набора, выводящую результат в виде ответа «ДА» или «НЕТ».

Формат входных данных: В первой строке подаётся число n (0 < n < 1000) – количество чисел в наборе. В 
последующих n строках вводятся целые числа, составляющие набор (могут повторяться). Затем следует целое 
число, которое является или не является произведением двух каких-то чисел из набора.

Формат выходных данных: Программа должна вывести «ДА» или «НЕТ» в соответствии с условием задачи.

Примечание 1. Само на себя число из набора умножиться не может, другими словами, два множителя должны иметь
разные индексы в наборе.

Примечание 2. Для решения задачи используйте вложенные циклы.
'''

def number_multiplication(numbers):
    number_set = []
    
    for i in range(numbers):
        number_set.append(int(input(f'Введите {i+1}-ое целое число набора: ')))
    
    l = number_set
    
    num = int(input('Введите проверяемое целое число: '))
    
    for n in range(len(number_set)):
        l = number_set[n+1:] # для исключения умножения числа на само себя (одинаковое число именно по индесу)
        #print(f'Список l {l}')
        
        for j in l:
            #print(f'{number_set[n]} * {j} = {number_set[n] * j}')
            
            if number_set[n] * j == num:
                return 'ДА, проверяемое число является произведением двух чисел из данного набора'
    
    return 'НЕТ, проверяемое число не является произведением двух чисел из данного набора'

print(number_multiplication(int(input('Введите количество чисел в наборе: '))))
```

```python
'''
Орел и решка. Дана строка текста, состоящая из букв русского алфавита "О" и "Р". Буква "О" соответствует
выпадению Орла, а буква "Р" - выпадению Решки. Напишите программу, которая подсчитывает наибольшее количество
подряд выпавших Решек.

Формат входных данных: На вход программе подается строка текста, состоящая из букв русского алфавита "О" и "Р".
Формат выходных данных: Программа должна вывести наибольшее количество подряд выпавших Решек.

Примечание. Если выпавших Решек нет, то необходимо вывести число 0.
'''

def heads_or_tails(text):
    text = [i for i in text]
    counter = 0
    l = []
    
    for i in text:
        if i == 'Р':
            counter += 1        
        else:
            counter = 0
        l.append(counter)
        
    return max(l)

print(heads_or_tails(input('Введите строку текста: ')))
```
